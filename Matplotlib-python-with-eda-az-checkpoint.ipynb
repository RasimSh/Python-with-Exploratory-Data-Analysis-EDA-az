```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 # Məqsəd yalnız qrafik çəkmək deyil, məlumatın strukturunu, paylanmasını və əlaqələrini vizual olaraq anlamaqdır. 
# Pandas, Matplotlib birlikdə istifadə olunanda həm sürətli prototipləşdirmə, həm də peşəkar vizualizasiya imkanı yaradır.

data = {
    "Ad": ["Tural", "Aysel", "Kamran", "Leyla"],
    "Yaş": [25, 30, 28, 22],
    "Maaş": [1200, 1500, 1100, 1000]
}

df = pd.DataFrame(data)
```

```python
# Pandas DataFrame üzərində birbaşa .plot() metodu ilə ilkin qrafiklər qurmaq mümkündür

# Bar chart
df["Yaş"].plot(kind="bar", title="Yaşların paylanması", color="steelblue")
plt.xlabel("İndeks")
plt.ylabel("Yaş")
plt.show()

# Line chart
df["Maaş"].plot(kind="line", marker="o", linestyle="--", title="Maaşların dinamikası")
plt.xlabel("İndeks")
plt.ylabel("Maaş")
plt.show()
```

```python
# Matplotlib daha geniş customization imkanı verir: rənglər, markerlər, annotasiyalar.

# Işçilərin yaş dinamikasını göstərən LineChart
plt.plot(df["Ad"], df["Yaş"], color="blue", marker="o", linestyle="--")
plt.title("İşçilərin Yaş Dinamikası")
plt.xlabel("Ad")
plt.ylabel("Yaş")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
# BarChart ilə maaş müqayisəsi
plt.bar(df["Ad"], df["Maaş"], color="orange", edgecolor="black")
plt.title("İşçilərin Maaş Müqayisəsi")
plt.xlabel("Ad")
plt.ylabel("Maaş")
plt.show()
```

```python
# 2 data arasındakı əlaqəni göstərən Scatter Plot
plt.scatter(df["Yaş"], df["Maaş"], c="darkred", s=100, alpha=0.7)
plt.title("Yaş və Maaş arasındakı əlaqə")
plt.xlabel("Yaş")
plt.ylabel("Maaş")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
# Bir və daha çox qrafiki eyni çərçivədə görmək
fig, axs = plt.subplots(1, 2, figsize=(10,4))

# Sol tərəfdə bar chart
axs[0].bar(df["Ad"], df["Maaş"], color="green")
axs[0].set_title("Maaş müqayisəsi")

# Sağ tərəfdə line chart
axs[1].plot(df["Ad"], df["Yaş"], color="blue", marker="o")
axs[1].set_title("Yaş dinamikası")

plt.tight_layout()
plt.show()
```
