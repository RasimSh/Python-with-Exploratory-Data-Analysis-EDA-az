```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# NumPy massivlərinin yaradılması (array, zeros, ones, arange, linspace).

# Python listi
siyahi = [10, 20, 30, 40]
# NumPy massivi
arr = np.array(siyahi)

print(f"Massin: {arr} \n")

# ---------------------------------------------------------------------------------------------------

# Sıfırlardan ibarət 2 sətir və 3 sütuna malik olan massiv.
arr_zeros = np.zeros((2, 3)) 
# 5 dene yan-yana birlər massiv. 1 sətir 5 sütun kimi görün.
arr_ones = np.ones(5)        

print("Sıfırlar matrisi:\n", arr_zeros)
print("Birlər massivi:", arr_ones, "\n")

# ---------------------------------------------------------------------------------------------------

# start, stop, step deyilen massiv.
# 0-da başlayır, 20-yə qədər davam edir, 3-3 artır. stop (yəni 20) daxil deyil.
arr_arange = np.arange(0, 20, 3)

print("3-3 artan sıra:", arr_arange, "\n")

# ---------------------------------------------------------------------------------------------------

# 0-dan başla, 1-də bitir (1 daxildir), ümumi 5 bərabər hissəyə bölür.
arr_linspace = np.linspace(0, 1, 5)

print("Bərabər bölünmüş hissələr:", arr_linspace, "\n")
```

```python
# - Statistik funksiyalar (mean, median, std).

# numeric datanın ümumi cəminin sayına olan nisbəti.
data = np.array([10, 20, 30, 40, 50]) 
orta = np.mean(data)
print(f"Orta qiymət: {orta} \n")

# ---------------------------------------------------------------------------------------------------

# datanın 50% böyük, qalan 50% nən kiçik olan rəqəm.
median_qiymet = np.median(data)
print(f"Median: {median_qiymet} \n")

# ---------------------------------------------------------------------------------------------------

# Rəqəmlərin orta qiymətdən nə qədər səpələndiyini göstərir.
# İki fərqli massiv yaradaq. Hər iki datanın medianı 50.
list_A = np.array([45, 50, 55]) 
list_B = np.array([1, 50, 99])  

print(f"A listinin səpələnməsi (std): {np.std(list_A)}")
print(f"B listinin səpələnməsi (std): {np.std(list_B)}")
print(f"Səpələnmə dərəcəsinə görə A listi üçün stabil = ({np.std(list_A)}), B listinə qeyri-stabil = ({np.std(list_B)}) ")
```

```python
# - Random modulundan istifadə (np.random.rand, np.random.randint).

# Bu funksiya 0 ilə 1 arasında bərabər paylanmış təsadüfi onluq ədədlər yaradır. 
# 2x2 ölçülü matris .
rand_matrix = np.random.rand(2, 2)
print("Təsadüfi 2x2 matris:\n", rand_matrix)

# ---------------------------------------------------------------------------------------------------

# (başlanğıc, son, say)
# 1-dən 100-ə qədər 5 dənə tam ədəd
tam_ededler = np.random.randint(1, 100, 5)

print("Təsadüfi tam ədədlər:", tam_ededler)
```
