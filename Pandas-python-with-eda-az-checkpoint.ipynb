```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# - Pandas python üçün verilənlərin analizi və manipulyasiyası kitabxanasıdır.
# - Əsas obyektləri: Series (1D) və DataFrame (2D cədvəl).
# - CSV, Excel, SQL, JSON kimi müxtəlif formatlardan verilənləri oxumaq və yazmaq imkanı verir.


# - CSV faylının oxunması
df_csv = pd.read_csv("data.csv")
print("CSV faylından oxunan ilk 5 sətir:")
print(df_csv.head())

# - Excel faylının oxunması
df_excel = pd.read_excel("data.xlsx", sheet_name=0)
print("\nExcel faylından oxunan ilk 5 sətir:")
print(df_excel.head())

# - JSON faylının oxunması
df_json = pd.read_json("data.json")
print("\nJSON faylından oxunan ilk 5 sətir:")
print(df_json.head())
```

```python
# - Sadə data dictionary
data = {
    "Ad": ["Tural", "Aysel", "Kamran", "Leyla"],
    "Yaş": [25, 30, 28, 22],
    "Şəhər": ["Bakı", "Gəncə", "Sumqayıt", "Şəki"],
    "Maaş": [1200, 1500, 1100, 1000]
}

# DataFrame yaratmaq
df = pd.DataFrame(data)

print("DataFrame:")
print(df)
```

```python
# - İlk sətirlərə baxmaq
print(f"{df.head(2)}\n")   

# - Sütun seçmək
print(f"{df["Ad"]}\n")     

# -  Şərt tətbiq etmək
print(f"{df[df["Yaş"] > 25]}\n")

# - Yeni sütun əlavə etmək
df["Bonus"] = df["Maaş"] * 0.1
print(f"{df}\n")

# - Sadə statistik göstəricilər 
# - Orta, min, max, std, quartiers
print(f"{df.describe()}\n")   

# - Qruplaşdırma - şəhərə görə qruplaşdırır.
print(f"{df.groupby("Şəhər")["Maaş"].mean()}\n")
```

```python
data = {
    "Ad": ["Tural", "Aysel", "Kamran", "Leyla", "Aysel"],
    "Yaş": [25, np.nan, 28, 22, 30],
    "Şəhər": ["Bakı", "Gəncə", "Sumqayıt", None, "Gəncə"],
    "Maaş": [1200, 1500, None, 1000, 1500]
}

df = pd.DataFrame(data)
print(f"DataFrame:\n {df}\n")

# - NaN dəyərlərin yoxlanması və cəmi tapmaq
print(f"{df.isnull().sum()}\n")

# - NaN dəyərlərin doldurulması- maaş sütununda NaN dəyərə mean verdi
df["Maaş"] = df["Maaş"].fillna(df["Maaş"].mean())
print(f"{df}\n")

# - Boş sətirlərin silinməsi - NaN olan sətirləri datadan silmək
# df = df.dropna()
print(f"{df}\n")

# -  Sütun adlarını dəyişmək
df = df.rename(columns={"Ad": "Name", "Yaş": "Age"})
print(df)
```

```python
# Bu boxda Pandas ilə DataFrame üzərində statistik əməliyyatlar göstərilir:
# - Orta, median, standart sapma
# - Minimum və maksimum dəyərlər
# - `describe()` ilə ümumi təsvir

data = {
    "Ad": ["Tural", "Aysel", "Kamran", "Leyla"],
    "Yaş": [25, 30, 28, 22],
    "Maaş": [1200, 1500, 1100, 1000]
}

df = pd.DataFrame(data)
print(f"{df}\n")

# - Orta qiymet (mean)
print(f"Orta yaş: {df["Yaş"].mean()}")
print(f"Orta maaş: {df["Maaş"].mean()}\n")

# - Median
print(f"Median yaş: {df["Yaş"].median()}\n")

# - Standart sapma
print(f"Yaşın standart sapması: {df["Yaş"].std()}\n")

# - Minimum və maksimum
print(f"Minimum maaş: {df["Maaş"].min()}")
print(f"Maksimum maaş: {df["Maaş"].max()} \n")

# - Datanın numerik qiymətləri üçün ümumi təsvir 
print(f"Ümumilikdə sadə statislik qiymətlər \n {df.describe()}\n")
```
