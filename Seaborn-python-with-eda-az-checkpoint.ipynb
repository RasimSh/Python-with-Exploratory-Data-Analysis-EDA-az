```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Seaborn — Python üçün statistik vizualizasiya kitabxanasıdır.
# Seaborn Matplotlib üzərində qurulmuş high-level (yüksək səviyyəli) vizualizasiya kitabxanasıdır.
# - rəng palitraları, mövzular (themes) və estetik dizayn təmin edir.
# - statistik verilənlərin paylanmasını, əlaqələrini və tendensiyalarını vizual şəkildə göstərmək.
data = {
    "Ad": ["Tural", "Aysel", "Kamran", "Leyla"],
    "Yaş": [25, 30, 28, 22],
    "Maaş": [1200, 1500, 1100, 1000]
}

df = pd.DataFrame(data)
```

```python
# Histogram və Kernel Density Estimation
# Kernel Density Estimation - paylanma sıxlığını təxmin etmə üsuludur. 
# Histogramdan fərqli olaraq, KDE verilənlərin paylanmasını davamlı və hamar əyri şəklində göstərir.

plt.figure(figsize=(6,4))
sns.histplot(df["Maaş"], bins=5, kde=True, color="skyblue")
plt.title("Maaşların Paylanması (Histogram + KDE)")
plt.show()
```

```python
# - EDA zamanı çox dəyişənli datasetlərdə əlaqələri tez görmək.
# - Korrelyasiya və outlierləri vizual aşkar etmək.
# - Model seçimi üçün ilkin təsəvvür yaratmaq (məsələn, linear regression üçün uyğunluq).

sns.pairplot(df[["Yaş", "Maaş"]], diag_kind="kde", plot_kws={"alpha":0.7})
plt.suptitle("Yaş və Maaş arasındakı əlaqələrin vizual analizi", y=1.02)
plt.show()
```

```python
# - Heatmap — verilənlərin matris şəklində vizualizasiyasıdır, rənglərlə dəyərlərin böyüklüyünü göstərir.
# - Ən çox istifadə olunduğu yer: korrelyasiya matrisi.
# - Rənglər vasitəsilə hansı dəyişənlərin bir‑birinə daha çox bağlı olduğunu tez görmək olur.

corr = df[["Yaş", "Maaş"]].corr()

plt.figure(figsize=(5,4))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Korrelyasiya Heatmap")
plt.show()
```

```python
# Seaborn ilə Subplot
# - Birdən çox qrafiki eyni pəncərədə göstərmək.
# - Müxtəlif dəyişənlərin paylanmasını və əlaqəsini yan‑yana müqayisə etmək.
# - EDA zamanı həm paylanma, həm də əlaqələri bir baxışda görmək üçün çox faydalıdır.

fig, axs = plt.subplots(1, 2, figsize=(12,5))

# Sol tərəfdə histogram
sns.histplot(df["Yaş"], bins=5, kde=True, ax=axs[0], color="lightgreen")
axs[0].set_title("Yaşların Paylanması")

# Sağ tərəfdə scatter plot
sns.scatterplot(x="Yaş", y="Maaş", data=df, hue="Ad", style="Ad", s=100, ax=axs[1])
axs[1].set_title("Yaş və Maaş əlaqəsi")

plt.tight_layout()
plt.show()
```
